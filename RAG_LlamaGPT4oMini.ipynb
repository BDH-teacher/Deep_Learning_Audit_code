```python
from langchain_unstructured import UnstructuredLoader
page_url = "https://www.samsunghospital.com/gb/language/english/guide/parking.do"
loader = UnstructuredLoader(web_url=page_url)

docs = []
async for doc in loader.alazy_load():
    docs.append(doc)
```

```python
for d in docs:
  print(d)
```

```python
from uuid import uuid4
uuids = [str(uuid4()) for _ in range(len(docs))]
print(uuids)
```

```python

from langchain_huggingface import HuggingFaceEmbeddings
model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
embedding_model = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

index = faiss.IndexFlatL2(len(embedding_model.embed_query("test")))

vector_store = FAISS(
    embedding_function=embedding_model,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
from langchain_core.documents import Document

vector_store.add_documents(documents=docs, ids=uuids)
```

```python
query = 'Do you have parking?'
results = vector_store.similarity_search(query,k=20)
contexts = ""
for i, result in enumerate(results):
    print(f"Result {i + 1}:")
    print(result)
    contexts += result.page_content
    contexts += " "
    print("\n---\n")
```

```python
import numpy as np
import faiss
from google.colab import drive
from langchain_huggingface import HuggingFaceEmbeddings

drive.mount("/content/drive")

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
embedding_model = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

loaded_index = faiss.read_index("/content/drive/MyDrive/faiss_index.index")
loaded_texts = np.load('/content/drive/MyDrive/faiss_texts.npy', allow_pickle=True)  # Path for texts mapping, allow_pickle=True)

query = 'Do you have parking?'
query_vector = embedding_model.embed_query(query)
print(query_vector)
query_vector = np.array(query_vector, dtype=np.float32)
if query_vector.ndim == 1:
    query_vector = query_vector.reshape(1, -1)
k=10
distances, indices = loaded_index.search(query_vector, k)
context = ""
for i in range(k):
    idx = indices[0][i]  # Get the index of the i-th nearest neighbor
    print(f"Text: {loaded_texts[idx]}")  # Retrieve the text using the index
    context += loaded_texts[idx].page_content
    context += " "
    print("\n---\n")
```

```python
from transformers import *
import huggingface_hub
from huggingface_hub.hf_api import HfFolder
HfFolder.save_token('YourToken')

model = AutoModelForCausalLM.from_pretrained("meta-llama/Meta-Llama-3-8B", torch_dtype="auto", load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3-8B", torch_dtype="auto")
tokenizer.use_default_system_prompt = False
```

```python
system_prompt = "You are a helpful assistant who is helping the Samsung Medical Center health system respond to patient questions. When you answer the question, please refer to the context, and use the first person (e.g., We) to refer to Samsung Medical Center."

# We generate contexts above
query = 'Do you have parking?'
def generate_response(query):
    input_text = f"""
    {system_prompt}\n\n\n

    Context: {context}\n\n\n

    Qustion: {query} \n
    Answer:
    """

    inputs = tokenizer(input_text, return_tensors="pt", truncation=True)
    outputs = model.generate(inputs["input_ids"], max_length=512,   # Maximum length of the output sequence
                            num_return_sequences=1,  # Number of sequences to return
                            do_sample=True,   # Whether to use sampling or greedy decoding
                            top_p=0.9,        # Nucleus sampling parameter
                            top_k=50,)         # Top-k sampling parameter
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = response.split("Answer:")[-1].strip()
    return response

print(generate_response(query))
```
