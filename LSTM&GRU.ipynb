```python
# PyTorch API - LSTM

import torch
import torch.nn as nn

# Define the LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        # Forward pass through the LSTM
        out, (h_n, c_n) = self.lstm(x)  # out: output features, h_n: hidden state, c_n: cell state
        return out, h_n, c_n


# Example usage
input_size = 10  # Number of input features per time step
hidden_size = 20  # Number of features in the hidden state
num_layers = 2  # Number of LSTM layers (stacked)

# Create LSTM model
model = LSTMModel(input_size, hidden_size, num_layers)

# Example input (batch_size, seq_length, input_size)
x = torch.randn(5, 15, input_size)

# Forward pass through the model
output, hidden, cell = model(x)
print(model)
print(output.shape)  # [batch_size, seq_length, hidden_size]
print(hidden.shape, cell.shape)
```

```python
# PyTorch API - GRU (사진 코드)

import torch
import torch.nn as nn

class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(GRUModel, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        # Forward pass through the GRU
        out, h_n = self.gru(x)  # out: output features, h_n: hidden state
        return out, h_n


# Example usage
input_size = 10  # Number of input features per time step
hidden_size = 20  # Number of features in the hidden state
num_layers = 2  # Number of GRU layers (stacked)

# Create GRU model
model = GRUModel(input_size, hidden_size, num_layers)

# Example input (batch_size, seq_length, input_size)
x = torch.randn(5, 15, input_size)

# Forward pass through the model
output, hidden = model(x)
print(model)
print(output.shape)  # [batch_size, seq_length, hidden_size]
print(hidden.shape)
```
