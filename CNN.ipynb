```python
torch.manual_seed(0)

x = torch.randn(1, 3, 32, 32)
conv = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5, stride=1)
pool = nn.MaxPool2d(kernel_size=2, stride=2)
relu = nn.ReLU()

conv_output = conv(x)
conv_output = relu(conv_output)
pooled_output = pool(conv_output)

print("Convolution Output Shape:", conv_output.shape)
print("Pooled Output Shape:", pooled_output.shape)
```
