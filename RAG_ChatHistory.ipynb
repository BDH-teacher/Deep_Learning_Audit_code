```python
from langchain_unstructured import UnstructuredLoader
page_url = "https://www.timeanddate.com/weather/south-korea/seoul/ext"
loader = UnstructuredLoader(web_url=page_url)

docs = []
async for doc in loader.alazy_load():
    docs.append(doc)
```

```python
for d in docs:
  print(d)
from uuid import uuid4
uuids = [str(uuid4()) for _ in range(len(docs))]
print(uuids)
```

```python

from langchain_huggingface import HuggingFaceEmbeddings
model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
embedding_model = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
!pip install -qU langchain-community faiss-cpu
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

index = faiss.IndexFlatL2(len(embedding_model.embed_query("test")))

vector_store = FAISS(
    embedding_function=embedding_model,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
from langchain_core.documents import Document

vector_store.add_documents(documents=docs, ids=uuids)
```

```python
def rag_retrieval(query):
  results = vector_store.similarity_search(query,k=20)
  contexts = ""
  for i, result in enumerate(results):
      contexts += result.page_content
      contexts += " "
  return contexts
```

```python
from openai import OpenAI
import os
os.environ['OPENAI_API_KEY']="YourKey"
client = OpenAI(
  api_key=os.environ['OPENAI_API_KEY']  # this is also the default, it can be omitted
)
GPT_MODEL = "gpt-4o-mini"
chat_history = []
def get_completion(prompt, contexts=None, model=GPT_MODEL):
    contexts += " ".join(chat_history)
    response = client.chat.completions.create(
        model=GPT_MODEL,
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant who is helping to answer everyday questions.",
            },
            {
                "role": "user",
                "content": prompt,
            },
            {
                "role": "assistant",
                "content": f"Use this following information as context to answer the user question: {contexts}. Please stick to this context when answering the question"
            },

        ],
    )
    features = response.choices[0].message.content
    chat_history.append(features)
    chat_history = chat_history[-3:]

    return features
```

```python

# Test the function with a prompt
query = "what's the weather in Seoul?"
rag_contexts = rag_retrieval(query)
res = get_completion(query, rag_contexts)
print(res)

query = 'How about next week?'
rag_contexts = rag_retrieval(query)
res = get_completion("How about next week?",rag_contexts)
print(res)
```
